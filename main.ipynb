```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data into pandas
df = pd.read_csv('Officer_JO Moral Alignment Ranking 2024 (Responses) - Form Responses 1.csv')
df.head()
```

```python
# clean up csv
df = df.drop(df.columns[:2], axis=1)
names = ['Alex Lu', 'Shawn Hatcher', 'Selina Wu', 'Jessica Yin', 'Raymond Sun', 'Naomi Norton', 'Cris Ferreon', 'Marcel Lee', 'Jayden Lu', 'Cindy Chai', 'Victoria Vuong', 'Cat Li', 'Jing Zhang', 'Zhikai Tang', 'Elaina Lan', 'Zoie Lee', 'Timmy Huang', 'Michael Ouyang', 'Zijing Lian', 'Aaron Hew', 'Andrew Lee', 'Shuhan Sun', 'Rohit Bandi', 'Nick Nguyen', 'Jessie Chen', 'Crystal Lee', 'Diana Yang', 'Josh Dong', 'Audrey Tan', 'Jordan Chen', 'Jason Li', 'Aaron Hu', 'Corn Wang', 'Diane Guo', 'Jely Lin', 'Annie Teng', 'Cyndi Liang']
columnLables = []
# Iterate over each name and generate two variations
for name in names:
    lawful_chaotic = name + " LawfulChaotic"
    evil_good = name + " EvilGood"
    columnLables.extend([lawful_chaotic, evil_good])
df.columns = columnLables
display(df)
```

```python

for x in range(len(names)):
    ax = df.plot.scatter(x=df.columns[x*2], y=df.columns[x*2+1],xlim=(0, 10), ylim=(0, 10),figsize=(6, 6))
    ax.axvline(x=5, color='black', linestyle='--')
    ax.axhline(y=5, color='black', linestyle='--')
    ax.set_xticks(range(11))
    ax.set_yticks(range(11))
    ax.grid(which='both')
    ax.set_xlabel(r'Lawful $\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\quad$ Chaotic')
    ax.set_ylabel(r'Evil $\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad\qquad$  Good')
    ax.set_title(names[x]+'\'s Moral Alignment Responses')
    plt.savefig(f'{names[x].replace(" ", "")}Responses.png', bbox_inches='tight')
```

```python
avg_df = pd.DataFrame(df.mean(), columns=['Average'])
display(avg_df)
even_rows = avg_df[::2]
odd_rows = avg_df[1::2]
even_rows.columns = ['Lawful - Chaotic']
odd_rows.columns = ['Evil - Good']
odd_rows.reset_index(drop=True, inplace=True)
even_rows.reset_index(drop=True, inplace=True)
# display(even_rows)
# display(odd_rows)
result_df = pd.concat([even_rows, odd_rows], axis=1)
result_df.index = names
display(result_df)
```

```python

result_df.loc['CSA'] = [4.946909,5.723713]
names.append("CSA")
ax = result_df.plot.scatter(x='Lawful - Chaotic', y='Evil - Good',xlim=(0, 10), ylim=(0, 10), figsize=(15, 15))

last_row = result_df.iloc[[-1]]
ax.scatter(last_row['Lawful - Chaotic'], last_row['Evil - Good'], color='red')

ax.axvline(x=5, color='black', linestyle='--')
ax.axhline(y=5, color='black', linestyle='--')
ax.set_xticks(range(11))
ax.set_yticks(range(11))
ax.grid(which='both')
xspace = '\qquad'*20
yspace = '\qquad'*20
ax.set_xlabel(rf'Lawful ${xspace}$ Chaotic')
ax.set_ylabel(rf'Evil ${yspace}$  Good')
ax.set_title('CSA JO Officerboard Moral Alignment')
for idx, row in result_df.iterrows():
    ax.annotate(idx.split()[0], (row['Lawful - Chaotic'], row['Evil - Good']), xytext=(5,-5),
                textcoords='offset points', family='sans-serif', fontsize=10)
ax.annotate('CSA', (row['Lawful - Chaotic'], row['Evil - Good']), xytext=(5,-5),
                textcoords='offset points', family='sans-serif', fontsize=10, color = 'red')
plt.savefig('CSAMoralAlignment.png',bbox_inches='tight')
r_value = result_df['Lawful - Chaotic'].corr(result_df['Evil - Good'])
print(r_value)
csa = result_df.mean()
print(csa)
```
